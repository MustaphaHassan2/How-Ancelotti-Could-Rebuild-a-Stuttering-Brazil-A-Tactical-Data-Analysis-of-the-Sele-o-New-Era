Defensive Foundations: Marquinhos as the New Rüdiger

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

labels = ['Tackles', 'Tackles Won', 'Def 3rd', 'Mid 3rd', 'Att 3rd',
          'Challenges Won', 'Challenges Attempted', 'Challenge %', 'Challenges Lost',
          'Blocks', 'Shot Blocks', 'Pass Blocks', 'Interceptions', 'Tkl+Int', 'Clearances']

rudiger_stats = [18, 11, 12, 6, 0, 10, 14, 71.4, 4, 21, 13, 8, 8, 26, 86]
marquinhos_stats = [30, 19, 8, 16, 6, 25, 33, 75.8, 8, 27, 15, 12, 11, 41, 64]

def normalize(stats):
    stats = np.array(stats)
    return stats / stats.max()

rudiger_norm = normalize(rudiger_stats).tolist()
marquinhos_norm = normalize(marquinhos_stats).tolist()

rudiger_norm += rudiger_norm[:1]
marquinhos_norm += marquinhos_norm[:1]
rudiger_stats += rudiger_stats[:1]
marquinhos_stats += marquinhos_stats[:1]

num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
fig.patch.set_facecolor('#0A002A')

gradient = np.linspace(0.07, 0.15, 100)
gradient = np.outer(gradient, gradient[::-1])
ax.imshow(gradient, extent=[-1, 1, -1, 1], transform=ax.transAxes,
          cmap='Greys', alpha=0.2, zorder=0)

ax.set_facecolor('#0A002A')

ax.plot(angles, rudiger_norm, color='white', linewidth=2, label='Rüdiger')
ax.fill(angles, rudiger_norm, color='white', alpha=0.15)

ax.plot(angles, marquinhos_norm, color='#002D72', linewidth=2, label='Marquinhos')
ax.fill(angles, marquinhos_norm, color='#ED1C24', alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, color='white', fontsize=10)
ax.set_yticklabels([])
ax.grid(color='white', alpha=0.1)
ax.spines['polar'].set_color('white')

fontstyle = {
    'fontsize': 10,
    'fontweight': 'bold',
    'fontname': 'Tahoma'
}

for angle, value, raw in zip(angles, rudiger_norm, rudiger_stats):
    ax.text(angle, value + 0.05, f'{raw:.0f}', ha='center', va='center', color='white', zorder=5, **fontstyle)

for angle, value, raw in zip(angles, marquinhos_norm, marquinhos_stats):
    ax.text(angle, value + 0.13, f'{raw:.0f}', ha='center', va='center', color="#D847478F", zorder=5, **fontstyle)

plt.title('Defensive Foundations: Marquinhos as the New Rüdiger', color='white', fontsize=15, weight='bold', pad=20)

legend = ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1), frameon=False, fontsize=12)
for text in legend.get_texts():
    text.set_color("white")

def add_tfa_logo(fig):
    logo_path = "D:\\Work\\Data Analysis TFA\\06 How Ancelotti Could Reshape Brazil\\logos\\TFA.png"
    tfa_logo = Image.open(logo_path).convert("RGBA")
    logo_width, logo_height = tfa_logo.size
    tfa_logo = tfa_logo.resize((int(logo_width * 0.257), int(logo_height * 0.257)))
    fig.figimage(tfa_logo, xo=29, yo=10, alpha=0.8, zorder=10)

add_tfa_logo(fig)

plt.tight_layout()
plt.show()
```

Full-Back Philosophy: The Mendy Template

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

def_labels = ['Tackles', 'Tackles Won', 'Def 3rd', 'Mid 3rd', 'Att 3rd',
              'Challenges Won', 'Challenges Attempted', 'Challenge %', 'Challenges Lost',
              'Blocks', 'Shot Blocks', 'Pass Blocks', 'Interceptions', 'Tkl+Int', 'Clearances']

mendy_def_stats = [32, 38, 17, 14, 7, 9, 12, 75, 3, 11, 4, 7, 9, 26, 8]

att_labels = ['Total Passes', 'Pass Completion %', 'Short Pass %', 'Medium Pass %', 'Long Pass %',
              'Progressive Passes', 'Progressive Carries', 'Progressive Receptions',
              'Total Distance', 'Progressive Distance', 'xA', 'xGChain', 'xG Buildup']

mendy_att_stats = [630, 93.9, 97, 93.9, 84.2, 36, 12, 40, 9091, 2431, 0.05, 0.09, 0]

def normalize(stats):
    stats = np.array(stats)
    return stats / stats.max()

def draw_radar(ax, labels, stats, title):
    norm_stats = normalize(stats).tolist()
    norm_stats += norm_stats[:1]
    stats += stats[:1]

    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    angles += angles[:1]

    ax.set_facecolor('#0A002A')
    ax.plot(angles, norm_stats, color='white', linewidth=2, label='Mendy')
    ax.fill(angles, norm_stats, color='white', alpha=0.15)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, color='white', fontsize=11, fontname='Tahoma')
    ax.set_yticklabels([])
    ax.grid(color='white', alpha=0.1)
    ax.spines['polar'].set_color('white')

    fontstyle = {'fontsize': 11, 'fontweight': 'bold', 'fontname': 'Tahoma'}
    for angle, value, raw in zip(angles, norm_stats, stats):
        ax.text(angle, value + 0.06, f'{raw:.0f}' if raw >= 1 else f'{raw:.2f}',
                ha='center', va='center', color='white', zorder=5, **fontstyle)

    ax.set_title(title, color='white', fontsize=15, fontname='Tahoma', weight='bold', pad=20)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10), subplot_kw=dict(polar=True))
fig.patch.set_facecolor('#0A002A')

gradient = np.linspace(0.07, 0.15, 100)
gradient = np.outer(gradient, gradient[::-1])
for ax in [ax1, ax2]:
    ax.imshow(gradient, extent=[-1, 1, -1, 1], transform=ax.transAxes,
              cmap='Greys', alpha=0.2, zorder=0)

draw_radar(ax1, def_labels, mendy_def_stats.copy(),
           'Ferland Mendy – Defensive Profile 2024/25')
draw_radar(ax2, att_labels, mendy_att_stats.copy(),
           'Ferland Mendy – Attacking & Passing Profile 2024/25')

def add_tfa_logo(fig):
    logo_path = "D:\\Work\\Data Analysis TFA\\06 How Ancelotti Could Reshape Brazil\\logos\\TFA.png"
    try:
        tfa_logo = Image.open(logo_path).convert("RGBA")
        logo_width, logo_height = tfa_logo.size
        tfa_logo = tfa_logo.resize((int(logo_width * 0.267), int(logo_height * 0.267)))
        fig.figimage(tfa_logo, xo=29, yo=10, alpha=0.8, zorder=10)
    except FileNotFoundError:
        print("Logo not found. Skipping logo overlay.")

add_tfa_logo(fig)

plt.tight_layout()
plt.show()
```
